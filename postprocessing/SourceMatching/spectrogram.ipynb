Examine the spectrogram of certain segments in postfiltered_2020-08-19_11:40:03_5

```python
# imports
import soundfile as sf
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# load data
data1,samplerate = sf.read('/Users/yihanhu/Desktop/CSE4223/Data/multiple source/postfiltered_2020-08-19_11_40_03_5.raw', 
                          channels=4, 
                          samplerate=16000,
                          subtype='PCM_16'
                         )
```

```python
# 1:06 - 1:07, Yihan's voice is duplicated. 
t1,t2 = 66,67
c0 = data1[t1*samplerate:t2*samplerate-1,0]
c1 = data1[t1*samplerate:t2*samplerate-1,1]
f, t, Cxx0 = signal.stft(c0, samplerate, nperseg=256)
f, t, Cxx1 = signal.stft(c1, samplerate, nperseg=256)

plt.figure(figsize = (25,8))
plt.subplot(121)
plt.pcolormesh(t, f, np.abs(Cxx0), shading='auto')
plt.title('STFT Magnitude Channel0 array5')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.subplot(122)
plt.pcolormesh(t, f, np.abs(Cxx1), shading='auto')
plt.title('STFT Magnitude Channel1 array5')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.show()
```
