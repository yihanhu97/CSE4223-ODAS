### Pushing dataframe as a whole to SQL server. Edit: Now pushing to a local sql database. Plan to push it to a sql database on the CSE lab server later.

https://www.dataquest.io/blog/sql-insert-tutorial/

```python
from sqlalchemy import create_engine
```

```python
# create sqlalchemy engine
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="root",
                               pw="rootpasswordgiven",
                               db="odas"))
```

```python
dataframe.to_sql('odas_data', con = engine, if_exists = 'append', chunksize = 1000)
```
